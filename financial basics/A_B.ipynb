```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import plotly.express as px
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../../data/full_data/full_cp.csv', index_col = 0)
df = df[['AAPL','NVDA']]
```

```python
for i in ['AAPL','NVDA']:
    df[f'{i}rt'] = np.log(df[i]) - np.log(df[i]).shift(1)
```

```python
df.drop(['AAPL','NVDA'], inplace = True, axis = 1)
```

```python
df.dropna(axis = 0,inplace = True)
df.head()
```

```python
fig = px.line(df)
fig.update_layout(title = {'text' : 'returns by period',
                                'x':0.5,
                                'xanchor' : 'center'},
                        template = 'plotly_dark',
                        xaxis_rangeslider_visible=True,
                        width=1000, height=500)
fig.show()
```

```python
fig = px.scatter(df, y = df['AAPLrt'], x = df['NVDArt'])
fig.update_traces(marker_size=5)
fig.update_layout(template = 'plotly_dark', width=750, height=750)
fig.show()
```

```python
X = df['AAPLrt'].to_numpy().reshape(-1,1)
y = df['NVDArt'].to_numpy().reshape(-1,1)
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
df.std()
```

```python
alpha, beta = lr.intercept_, lr.coef_
alpha, beta
```
