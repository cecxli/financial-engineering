```python
import pandas as pd
import numpy as np
import yfinance as yf
import os
import glob
import plotly.express as px
import plotly.graph_objects as go
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../../data/full_data/full_df.csv')
```

```python
googl = df[df['Name'] == 'GOOGL']
googl['return'] = googl['Close'] / googl['Close'].shift(1) - 1
googl['pct_change'] = googl['Close'].pct_change(1)
googl['log_return'] = np.log(googl['Close']) - np.log(googl['Close']).shift(1)
googl.head()
```

```python
from plotly.subplots import make_subplots
fig = make_subplots(subplot_titles=['pct change', 'log return'],
                    rows=2, cols=1)

fig.add_trace(go.Histogram(x=googl['pct_change']), row=1, col=1)
fig.add_trace(go.Histogram(x=googl['log_return']), row=2, col=1)

fig.update_layout(showlegend=False, height=600, width=800)
fig.show()
```
