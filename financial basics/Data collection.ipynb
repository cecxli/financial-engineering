```python
import yfinance as yf
import os
import pandas as pd
import glob
import plotly.express as px
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
symbols = ['AAPL','ADBE','AMZN','ASML','AVGO',
        'CMCSA','COST','CSCO','FB','GOOG',
        'GOOGL','INTC','MSFT','NFLX','NTES',
        'NVDA','PEP','PYPL','TSLA'
        ]
```

```python
if not os.path.exists('../../data'):
  os.mkdir('../../data')
```

```python
files = glob.glob('../../data/*.csv')
def delete_csv(files):
    for f in files:
        os.remove(f)
```

```python
delete_csv(files)
```

```python
def download_data(symbols):
    for i in symbols:
        if not os.path.exists(f'../../data/{i}.csv'):
            data = yf.download(i,start = '2010-01-01')
            if data.size > 0:
                print(f'saving {i} as csv')
                data.to_csv(f'../../data/{i}.csv')
            else:
                print(f'could not save {i}')
```

```python
download_data(symbols)
```

```python
files = glob.glob('../../data/*.csv')
```

```python
for symbol in symbols:
  s = open(f"../../data/{symbol}.csv").readlines()
  if len(s) < 10:
    os.system(f"rm../../data/{symbol}.csv")
```

```python
full_df = None
for f in files:
  print(f)
  df = pd.read_csv(f)

  symbol = f.split('d')[1].split('.')[0]
  df['Name'] = symbol[4:]

  if full_df is None:
    full_df = df
  else:
    full_df = full_df.append(df, ignore_index=True)
if not os.path.exists('../../data/full_data'):
  os.mkdir('../../data/full_data')
full_df.to_csv('../../data/full_data/full_df.csv', index=False)
```

```python
full_df
```

```python
full_df['Date'] = pd.to_datetime(full_df['Date'])
```

```python
full_df['Date'].min(), full_df['Date'].max()
```

```python
dates = pd.date_range(full_df['Date'].min(), full_df['Date'].max())
```

```python
close_prices = pd.DataFrame(index=dates)
```

```python
symbols = full_df['Name'].unique()
symbols
```

```python
for i in symbols:
  df_sym = full_df[full_df['Name'] == i]
  df_tmp = pd.DataFrame(data=df_sym['Close'].to_numpy(), \
                        index=df_sym['Date'], columns=[i])
  close_prices = close_prices.join(df_tmp) 
```

```python
close_prices.dropna(inplace=True)
```

```python
close_prices
```

```python
def plotly_stock_prices(dataframe):
    print(f'plotting between dates \n{dataframe.index.values.min()}\n{dataframe.index.values.max()}')
    fig = px.line(dataframe)
    fig.update_layout(title = {'text' : 'stock prices by period',
                                'x':0.5,
                                'xanchor' : 'center'},
                        xaxis_title = 'Period slicer',
                        yaxis_title = 'Price',
                        template = 'plotly_dark',
                        xaxis_rangeslider_visible=True,
                        width=950, height=500)

    fig.show()
```

```python
plotly_stock_prices(close_prices)
```
